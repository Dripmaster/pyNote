```python
import numpy as np
import imageio
im = imageio.imread('img/3.jpg')
```

```python
reim = np.zeros((im.shape[0],im.shape[1]*2,im.shape[2]))
Rim = np.zeros((im.shape[1],im.shape[0],im.shape[2]))
```

```python
j = 0
i = 0
for j in range(reim.shape[0]):
    for i in range(reim.shape[1]):
        reim[j][i] = im[j][i//2]
j = 0
i = 0
for j in range(Rim.shape[0]):
    for i in range(Rim.shape[1]):
        Rim[j][i] = im[i][j]
reim = np.uint8(reim)
Rim = np.uint8(Rim)
```

```python
imageio.imwrite('img/3-resize.jpg', reim)
imageio.imwrite('img/3-rot.jpg', Rim)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.subplot(1, 3, 1)
plt.imshow(im)
plt.subplot(1, 3, 2)
plt.imshow(reim)
plt.subplot(1, 3, 3)
plt.imshow(Rim)
```
